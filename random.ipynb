```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
np.random.seed(0)

```

```python
#create an object called iris with iris data
iris=load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
  #add new column for the species name
df['species']=pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
#create test and train data
df['is_tarin']=np.random.uniform(0,1,len(df)) <= .75
df.head()
```

```python
#create dataframes with test rows and training rows
train,test=df[df['is_tarin']==True], df[df['is_tarin']==False]
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:', len(test))

```

```python
#create a list of features columns names
features=df.columns[:4]
print(features) 
```

```python
#convert each species name into digits
y=pd.factorize(train['species'])[0]
y
```

```python
#creating random forest classifier
clf=RandomForestClassifier(n_jobs=2, random_state=0,)
#training classifier
clf.fit(train[features], y)
```

```python
#apply the trained classifier to test
clf.predict(test[features])
```

```python
test[features]
```

```python
#view predicted probabilities of the first 10 observation
clf.predict_proba(test[features][10:20])
```

```python
#mapping names for the plants
preds=iris.target_names[clf.predict(test[features])]
preds[0:5]
```

```python
#view actual species
test['species'].head()
```

```python
#confusion matris
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```
